```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
data = pd.read_csv('Train.csv')
data['Age_Outlet'] = datetime.now().year - data['Outlet_Establishment_Year']
data.drop(columns=['Outlet_Establishment_Year', 'Outlet_Identifier', 'Item_Identifier'], inplace=True)
```

```python
data.sample(5)
```

### Numerical - Numerical 

```python
sns.scatterplot(x=data['Item_Outlet_Sales'], y=data['Item_Weight'])
```

* Not getting any relation between item_weight and item_sales

```python
sns.scatterplot(data['Item_Outlet_Sales'],data['Item_Visibility'])
```

* There are some Item who has low sales if there visibility is more than 20%

```python
sns.scatterplot(data['Item_Outlet_Sales'],data['Item_MRP'])
```

* Item having MRP of 150 or more are getting sales more than 7000.

```python
data = pd.read_csv('Train.csv')
# we need original year data for line plot
sns.lineplot(data['Outlet_Establishment_Year'], data['Item_Outlet_Sales'])
```

### Numerical - Categorical

```python
sns.barplot(data['Item_Fat_Content'], data['Item_Outlet_Sales'])
```

* Regular have slight more sales than Low Fat, we can also say both are doing good in sales.

```python
from matplotlib.pyplot import figure
figure(figsize=(21,12), dpi=80)
sns.barplot(data['Item_Type'], data['Item_Outlet_Sales'])
plt.grid(color='black', linewidth=3, axis='both', alpha=0.1)
```

* Starchy Food have the highest sales then Seafood is on number second on sales.
* Vegetables & Fruits and Snack Foods have the similar sales arround 2200.
* Housholds and Dairy share the kind of similar sales.
